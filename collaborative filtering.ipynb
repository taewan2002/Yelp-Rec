```python
def get_info_for_popular_restaurant(data):
    # ê°€ìž¥ ì¸ê¸° ìžˆëŠ” ë ˆìŠ¤í† ëž‘ì˜ business_id ê°€ì ¸ì˜¤ê¸°
    most_popular_business_ids = data.groupby('business_id')['stars'].count().sort_values(ascending=False).head().index
    
    restaurant_info = []
    
    for business_id in most_popular_business_ids:
        # í•´ë‹¹ business_idì— ëŒ€í•œ í•„í„°ë§
        Filter = data['business_id'] == business_id

        # ë ˆìŠ¤í† ëž‘ì˜ ì´ë¦„ê³¼ ì£¼ì†Œ ì¶œë ¥
        name = data[Filter]['name'].unique()
        address = data[Filter]['address'].unique()
        
        restaurant_info.append((name, address))

    return restaurant_info

# ì˜ˆì‹œ: ê°€ìž¥ ì¸ê¸° ìžˆëŠ” ë ˆìŠ¤í† ëž‘ì˜ ì •ë³´ ì¶œë ¥
restaurant_info = get_info_for_popular_restaurant(combined_business_data)

print("\nMost Popular Restaurants:")
for i, (name, address) in enumerate(restaurant_info, 1):
    print(f"\nRestaurant {i}:")
    print("Name:", name)
    print("Address:", address)
```

```python
# Building a Utility Matrix (User-Restaurant Matrix)
# ì‚¬ìš©ìž, ê° ë ˆìŠ¤í† ëž‘, ì‚¬ìš©ìžê°€ ë¶€ì—¬í•œ ë“±ê¸‰ í¬í•¨í•œ matrix

# create a user-item matrix
rating_crosstab = combined_business_data.pivot_table(values='stars', index='user_id', columns='name', fill_value=0)
rating_crosstab.head()
```

```python
# Transposing the Matrix
# ì‚¬ìš©ìž - column
# ë ˆìŠ¤í† ëž‘ - row
# shape of the Utility matrix (original matrix) 
rating_crosstab.shape
```

```python
# Transpose the Utility matrix
X = rating_crosstab.values.T
X.shape
```

```python
# Decomposing the Matrix
'''
Now we use TruncatedSVD from sklearn to compress the transposed matrix into down to a number of rows by 12 matrices. 
All of the restaurants are in the rows. 
But the users will be compressed down to 12 components arbitrarily that represent a generalized view of users' tastes.
'''
import sklearn
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics import accuracy_score


SVD = TruncatedSVD(n_components=12, random_state=17)
result_matrix = SVD.fit_transform(X)
result_matrix.shape
```

```python
from sklearn.decomposition import PCA

# PCA ëª¨ë¸ì„ ìƒì„±í•˜ê³  ë°ì´í„°ì— fit_transformì„ ì ìš©
pca = PCA()
result_matrix2 = pca.fit_transform(result_matrix)

# ì„¤ëª…ëœ ë¶„ì‚°ì˜ ëˆ„ì  ë¹„ìœ¨ í™•ì¸
cumulative_explained_variance_ratio = np.cumsum(pca.explained_variance_ratio_)

thresholds = []
selected_components = []

# 0.95ë¶€í„° 0.99ê¹Œì§€ì˜ ëˆ„ì  ì„¤ëª…ëœ ë¶„ì‚°ì˜ ëª©í‘œ ë¹„ìœ¨ì— ëŒ€í•´ ì£¼ì„±ë¶„ì˜ ê°œìˆ˜ ì„ íƒ ë° ì¶œë ¥
for threshold in np.arange(0.59, 0.99, 0.01):
    selected_n_components = (cumulative_explained_variance_ratio >= threshold).argmax() + 1
    thresholds.append(threshold)
    selected_components.append(selected_n_components)
    print(f"Threshold: {threshold}, ì„ íƒëœ ì£¼ì„±ë¶„ì˜ ê°œìˆ˜: {selected_n_components}")
```

```python
plt.plot(thresholds, selected_components, marker='o')
plt.title('Threshold vs Selected Components')
plt.xlabel('Threshold')
plt.ylabel('Selected Components')
plt.grid(True)
plt.show()
```

```python
def recommend_restaurants(user_id, rating_crosstab, corr_matrix, num_recommendations=5):
    # ì‚¬ìš©ìžê°€ ì´ë¯¸ í‰ê°€í•œ ë ˆìŠ¤í† ëž‘ ê°€ì ¸ì˜¤ê¸°
    user_ratings = rating_crosstab.loc[user_id]

    # ì‚¬ìš©ìžê°€ í‰ê°€í•œ ë ˆìŠ¤í† ëž‘ì„ ì œì™¸í•œ ìœ ì‚¬í•œ ë ˆìŠ¤í† ëž‘ ì°¾ê¸°
    similar_restaurants = pd.Series(dtype=float)
    for i, rating in user_ratings.iteritems():
        similar_restaurants = similar_restaurants.append(corr_matrix[i].drop(i).map(lambda x: x * rating))

    # ìœ ì‚¬ë„ì— ë”°ë¼ ë ˆìŠ¤í† ëž‘ ì •ë ¬ í›„ ìƒìœ„ nê°œ ì„ íƒ
    similar_restaurants = similar_restaurants.sort_values(ascending=False).head(num_recommendations)

    # ì¶”ì²œëœ ë ˆìŠ¤í† ëž‘ ì •ë³´ ì¶œë ¥
    recommended_restaurants = []
    for restaurant_name, correlation in similar_restaurants.iteritems():
        recommended_restaurants.append((restaurant_name, correlation))

    return recommended_restaurants
```

```python
def calculate_hit_rate(data, corr_matrix, restaurant_names, n_components=12):
    """
    ì „ì²´ ë°ì´í„°ì…‹ì— ëŒ€í•œ Hit Rate ê³„ì‚° í•¨ìˆ˜
    :param data: ì „ì²´ ë°ì´í„°ì…‹ (DataFrame)
    :param corr_matrix: ìƒê´€ í–‰ë ¬
    :param restaurant_names: ë ˆìŠ¤í† ëž‘ ì´ë¦„ ëª©ë¡
    :param n_components: Truncated SVDì˜ ì»´í¬ë„ŒíŠ¸ ìˆ˜
    :return: ì „ì²´ ë°ì´í„°ì…‹ì— ëŒ€í•œ í‰ê·  Hit Rate ë° ê° ì‚¬ìš©ìžì˜ ì¶”ì²œëœ ë ˆìŠ¤í† ëž‘ ëª©ë¡
    """
    hit_rates = []
    user_recommendations = {}  # Dictionary to store recommended items for each user

    # ì‚¬ìš©ìž-ë ˆìŠ¤í† ëž‘ í‰ì  í‘œ ìƒì„±
    rating_crosstab = data.pivot_table(values='stars', index='user_id', columns='name', fill_value=0)
    X = rating_crosstab.values.T

    # Truncated SVDë¥¼ ì‚¬ìš©í•œ ì°¨ì› ì¶•ì†Œ
    SVD = TruncatedSVD(n_components=n_components, random_state=17)
    result_matrix = SVD.fit_transform(X)

    # ìƒê´€ í–‰ë ¬ ê³„ì‚°
    corr_matrix = np.corrcoef(result_matrix)

    # user_idë¥¼ ìˆ«ìžë¡œ ë§¤í•‘
    user_id_mapping = {user_id: i for i, user_id in enumerate(data['user_id'].unique())}

    for user_id in data['user_id'].unique():
        user_index = user_id_mapping[user_id]

        # ì‚¬ìš©ìžê°€ í‰ê°€í•œ ë ˆìŠ¤í† ëž‘ì„ ì œì™¸í•˜ê³  ìœ ì‚¬í•œ ë ˆìŠ¤í† ëž‘ ì°¾ê¸°
        if user_index >= len(corr_matrix):
            continue
        corr_user = corr_matrix[user_index, :]
        selected_restaurants = list(restaurant_names[(corr_user < 1.0) & (corr_user > 0.9)])

        # ëª¨ë¸ì´ ì¶”ì²œí•œ ì•„ì´í…œ ëª©ë¡
        recommended_items = selected_restaurants[:5]  # ì˜ˆì‹œë¡œ ìƒìœ„ 5ê°œ ì¶”ì²œ
        user_recommendations[user_id] = recommended_items

        # ì‹¤ì œ ì„ í˜¸í•˜ëŠ” ì•„ì´í…œ ëª©ë¡
        actual_items = data[data['user_id'] == user_id]['name'].tolist()

        # Hit Rate ê³„ì‚°í•˜ì—¬ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
        hr = hit_rate(recommended_items, actual_items)
        hit_rates.append(hr)

    return np.mean(hit_rates), user_recommendations

# HR ë° ì¶”ì²œëœ ë ˆìŠ¤í† ëž‘ ëª©ë¡ ê³„ì‚°
hr, user_recommendations = calculate_hit_rate(combined_business_data, corr_matrix, restaurant_names)
print(f"Hit Rate: {hr}")
print("User Recommendations:")
for user_id, recommendations in user_recommendations.items():
    print(f"User {user_id}: {recommendations}")
```
